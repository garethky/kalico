# PA Calibration Viewer

Loads a `.jsonl` file produced by `LOAD_CELL_PA_SWEEP_CAPTURE` and plots the force/time trace with junction markers for each run.

```python
from pathlib import Path

JSONL_PATH = Path("../pa_sweep_8.jsonl")
```

```python
import json

import matplotlib.pyplot as plt
import numpy as np

records = []
with JSONL_PATH.expanduser().open("r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if line:
            records.append(json.loads(line))

for i, rec in enumerate(records):
    time = np.asarray(rec["time"], dtype=float)
    force = np.asarray(rec["force"], dtype=float)
    moves = rec.get("trapq_moves", [])

    fig, ax = plt.subplots(figsize=(30, 5))
    ax.plot(time, force, linewidth=0.8, alpha=0.9)

    for mv in moves:
        a = mv["accel"]
        if a > 0:
            color = "green"
        elif a < 0:
            color = "red"
        else:
            color = "gray"
        ax.axvline(mv["print_time"], color=color, linewidth=0.6, alpha=0.7)

    speed = rec.get("speed", "?")
    accel = rec.get("accel", "?")
    advance = rec.get("advance", "?")
    flow = rec.get("flow_rate", "?")
    ax.set_title(f"Run {i}: speed={speed} accel={accel} advance={advance} flow={flow}")
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Force (g)")
    ax.grid(True, alpha=0.3)
    fig.tight_layout()
    plt.show()
```
